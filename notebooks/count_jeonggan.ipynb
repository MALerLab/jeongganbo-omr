```python
import cv2
reader = JeongganboReader()
title_png_dir = Path('title_pngs/')
title_png_dir.mkdir(exist_ok=True)

for instrument in instruments[4:]:
  num_titles = 0
  print(f"Parsing {instrument}...")
  png_fns = sorted(list(png_dir.glob(f'{instrument}_pg-*.png')))
  for fn in png_fns[page_starts_by_instrument[instrument]:]:
    img = reader.detect_title_box_from_img_path(str(fn))
    if img is not None:
      save_path = title_png_dir / (fn.stem + piece_name_by_instrument[instrument][num_titles] + '.png')
      # save image
      cv2.imwrite(str(save_path), img)
      num_titles += 1

```
